```python
from sklearn.metrics import roc_curve, roc_auc_score
import matplotlib.pyplot as plt

# 1) Probabilidades predichas por cada modelo
scores = {
    "Logit":  logit_model.predict_proba(X_test_2025)[:, 1],
    "KNN (K-CV)": knn_cv_model.predict_proba(X_test_2025)[:, 1],
    "Logit-LASSO": lasso_model.predict_proba(X_test_2025)[:, 1],
    "Logit-Ridge": ridge_model.predict_proba(X_test_2025)[:, 1],
}

# 2) Gráfico ROC
plt.figure(figsize=(7,6))
for name, sc in scores.items():
    fpr, tpr, _ = roc_curve(y_test_2025, sc)
    auc = roc_auc_score(y_test_2025, sc)
    plt.plot(fpr, tpr, label=f"{name} (AUC = {auc:.3f})")

# Línea diagonal = azar
plt.plot([0,1], [0,1], "k--", label="Azar (AUC = 0.5)")

plt.title("Curvas ROC - Modelos de Clasificación (X_test_2025)")
plt.xlabel("Tasa de falsos positivos (1 - Especificidad)")
plt.ylabel("Tasa de verdaderos positivos (Sensibilidad)")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import (
    accuracy_score, precision_score, recall_score,
    f1_score, roc_auc_score
)

rows = []
for name, sc in scores.items():
    y_pred = (sc >= 0.5).astype(int)
    rows.append({
        "Modelo": name,
        "F1": f1_score(y_test_2025, y_pred),
        "AUC": roc_auc_score(y_test_2025, sc),
    })

tabla_metricas = pd.DataFrame(rows).set_index("Modelo").round(3).sort_values("AUC", ascending=False)
display(tabla_metricas)
```
