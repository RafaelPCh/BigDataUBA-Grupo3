```python
def buscar_variable(lista_dfs, nombres, variable):
    """
    Busca si una variable está en una lista de DataFrames.

    Parámetros:
    -----------
    lista_dfs : list
        Lista de DataFrames de pandas.
    nombres : list
        Lista de nombres (strings) asociados a cada DataFrame.
    variable : str
        Nombre de la variable/columna a buscar.

    Retorna:
    --------
    dict : Diccionario con nombre del DataFrame y True/False si la variable está presente.
    """
    resultados = {}
    for df, nombre in zip(lista_dfs, nombres):
        resultados[nombre] = variable in df.columns
    return resultados

lista = [disc_207,disc_208,disc_209,disc_209,disc_210,disc_211_1,disc_211_2,disc_211_3,disc_213,disc_212,disc_214,disc_211_2,disc_215,disc_216]
nombres = ["disc_207","disc_208","disc_209","disc_209","disc_210","disc_211_1","disc_211_2","disc_211_3","disc_213","disc_212","disc_214","disc_211_2","disc_215","disc_216"]
print(buscar_variable(lista, nombres, "P01"))




```

```python
def contar_filas_bases(bases: dict):
    """
    Recibe un diccionario con nombre_de_base: DataFrame
    y devuelve/imprime la cantidad de filas de cada uno.
    """
    for nombre, df in bases.items():
        print(f"{nombre}: {len(df)} filas")
```

```python
import matplotlib.pyplot as plt

# Serie auxiliar sin modificar el dataset original
genero_labels = disc_207["genero"].map({1: "Hombre", 2: "Mujer"})

# Calcular porcentajes
porcentajes = genero_labels.value_counts(normalize=True) * 100

# Diccionario de colores fijo
color_dict = {"Hombre": "#66b3ff", "Mujer": "#ff9999"}
colores = [color_dict[g] for g in porcentajes.index]

# Gráfico de torta
fig, ax = plt.subplots(figsize=(6, 6))
wedges, texts, autotexts = ax.pie(
    porcentajes,
    labels=None,  # sin etiquetas en la torta
    autopct="%1.1f%%",
    startangle=90,
    colors=colores,
    wedgeprops={"edgecolor": "white", "linewidth": 1.5},
    textprops={"fontsize": 12}
)

# Estilo de los porcentajes
for autotext in autotexts:
    autotext.set_color("black")
    autotext.set_fontsize(12)
    autotext.set_weight("bold")

# Título
plt.title("Distribución de Género", fontsize=16, weight="bold", pad=20)

# Leyenda al costado superior derecho
ax.legend(
    wedges,
    porcentajes.index,
    loc="upper right",
    bbox_to_anchor=(1.2, 1),
    fontsize=12,
    title_fontsize=13
)

# Mantener proporción circular
ax.axis("equal")

plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl

genero_labels = disc_207["genero"].map({1: "Hombre", 2: "Mujer"})
rango = pd.cut(
    disc_207["edad"],
    bins=[0, 12, 24, 39, 64, 150],
    labels=["Niños", "Adolescentes", "Jóvenes", "Adultos", "Mayores"]
)
conteo = pd.crosstab(rango, genero_labels).astype(int)

# ===== Colores =====
fondo = "#fcfbf6"       # beige suave
color_hombre = "#7aa6c2"  # azul grisáceo suave
color_mujer = "#e69a8d"   # coral apagado

# ===== Gráfico =====
fig, ax = plt.subplots(figsize=(8, 5), facecolor=fondo)
ax.set_facecolor(fondo)

conteo.plot(
    kind="bar",
    stacked=True,
    color=[color_hombre, color_mujer],
    ax=ax,
    edgecolor="white"
)

# Eje X y Y
ax.set_xlabel("")   # elimina el label "Rango etario"
ax.set_ylabel(None) # elimina el label del eje Y
plt.xticks(rotation=0, color="#222222")  # etiquetas horizontales
plt.yticks(color="#222222")
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)

# Leyenda
ax.legend(title="Género", fontsize=11, title_fontsize=12, frameon=False)
plt.savefig("grafico_genero.png", dpi=300, bbox_inches="tight")

plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# 1) Filtrar NAs
df_limpio = disc_207.dropna(subset=["departamento", "genero"]).copy()

# 2) Mapear género a texto SIN modificar el dataset
genero_labels = df_limpio["genero"].map({1: "Hombre", 2: "Mujer"})

# 3) Tabla de conteo
conteo = pd.crosstab(df_limpio["departamento"], genero_labels)
conteo = conteo.reindex(columns=["Hombre", "Mujer"], fill_value=0)

# 4) Convertir a porcentaje por departamento
totales = conteo.sum(axis=1)
conteo_porcentaje = conteo.div(totales.replace(0, np.nan), axis=0) * 100

# 5) Plot con fondo beige
fondo = "#fcfbf6"
fig, ax = plt.subplots(figsize=(14, 6), facecolor=fondo)
ax.set_facecolor(fondo)

# Colores
color_hombre = "#7aa6c2"
color_mujer  = "#e69a8d"

# Graficar con barras más anchas (width=0.7 o 0.8 en vez de 0.5 default)
conteo_porcentaje.plot(
    kind="bar",
    stacked=True,
    color=[color_hombre, color_mujer],
    ax=ax,
    edgecolor="white",
    linewidth=0.9,
    width=1
)

# Quitar eje Y
ax.yaxis.set_visible(False)

# Títulos y formato
ax.set_xlabel("")
plt.xticks(rotation=45, ha="right")

# Agregar etiquetas de porcentaje dentro de cada barra
for i, (h, m) in enumerate(zip(conteo_porcentaje["Hombre"], conteo_porcentaje["Mujer"])):
    if h > 0:
        ax.text(i, h/2, f"{h:.1f}%", ha="center", va="center", color="white", fontsize=9, fontweight="bold")
    if m > 0:
        ax.text(i, h + m/2, f"{m:.1f}%", ha="center", va="center", color="white", fontsize=9, fontweight="bold")
ax.get_legend().remove()
ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_visible(False)

plt.tight_layout()
plt.show()
```
